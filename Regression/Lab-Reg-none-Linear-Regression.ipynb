```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
org_df = pd.read_csv("china_gdp.csv")
print("head:")
print(org_df.head())
print("tail:")
print(org_df.tail())
```

```python
org_df.describe()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(org_df["Year"].values, org_df["Value"].values, "bo")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = 1 + 1/(1 + np.exp(-x))

plt.plot(x, y)
plt.ylabel("Dependent Variable")
plt.xlabel("Independent Variable")
plt.show()
```

```python
def sigmoid(x, Beta_1, Beta_2):
    y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
    return y
```

```python
beta_1 = 0.1
beta_2 = 2011

y_pred = sigmoid(org_df["Year"].values, beta_1, beta_2)

plt.plot(org_df["Year"].values, org_df["Value"].values, "bo")
plt.plot(org_df["Year"].values, y_pred*10000000000000, 'r-')
plt.show()
print(y_pred)
```

```python
x_data = org_df["Year"].values / max(org_df["Year"].values)
y_data = org_df["Value"].values / max(org_df["Value"].values)
```

```python
from scipy.optimize import curve_fit
popt, pcov  = curve_fit(sigmoid, x_data, y_data)
print("Beta_1:{0}, Beta_2:{1}".format(popt[0], popt[1]))
```

```python
x = np.linspace(1960, 2014, 55)
x = x/max(x)
y = sigmoid(x, *popt)
plt.figure(figsize=(10, 5))
plt.plot(x_data, y_data, "bo", label="data")
plt.plot(x, y, "r", label="fit")
plt.legend(loc="best")
plt.show()
```
