```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
org_df = pd.read_csv("FuelConsumption.csv")
print('shape of org_df:',  org_df.shape)
print(org_df.dtypes)
org_df.head()
```

```python
org_df.describe()
```

```python
cdf = org_df[["ENGINESIZE", "CYLINDERS", "FUELCONSUMPTION_CITY", "FUELCONSUMPTION_HWY", "FUELCONSUMPTION_COMB", "CO2EMISSIONS"]]
cdf.head()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(10,10))

axs[0, 0].scatter(cdf["CO2EMISSIONS"], cdf["CYLINDERS"])
axs[0, 0].set_xlabel("CO2 Emissions")
axs[0, 0].set_ylabel("Cylinders")

axs[0, 1].scatter(cdf["CO2EMISSIONS"], cdf["FUELCONSUMPTION_CITY"])
axs[0, 1].set_xlabel("CO2 Emissions")
axs[0, 1].set_ylabel("Fuel Consumption City")

axs[1, 0].scatter(cdf["CO2EMISSIONS"], cdf["FUELCONSUMPTION_HWY"])
axs[1, 0].set_xlabel("CO2 Emissions")
axs[1, 0].set_ylabel("Fuel Consumption HWY")

axs[1, 1].scatter(cdf["CO2EMISSIONS"], cdf["FUELCONSUMPTION_COMB"])
axs[1, 1].set_xlabel("CO2 Emissions")
axs[1, 1].set_ylabel("Fuel Consumption Comb")

axs[2, 0].scatter(cdf["CO2EMISSIONS"], cdf["ENGINESIZE"])
axs[2, 0].set_xlabel("CO2 Emissions")
axs[2, 0].set_ylabel("Engine Size")
```

```python
msk = np.random.rand(len(cdf)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
x_train = np.asanyarray(train[["ENGINESIZE"]])
y_train = np.asanyarray(train[["CO2EMISSIONS"]])
reg.fit(x_train, y_train)

print("Coefficients:", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
plt.scatter(x_train, y_train, color = "black")
plt.plot(x_train, reg.intercept_[0] + x_train * reg.coef_[0][0], "-r")
plt.show()
```

```python
from sklearn.metrics import r2_score

x_test = np.asanyarray(test[["ENGINESIZE"]])
y_test = np.asanyarray(test[["CO2EMISSIONS"]])
y_ = reg.predict(x_test)

print("Mean absolute error: %.2f" % np.mean(np.absolute(y_ - y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_ - y_test) ** 2))
print("R2-score: %.2f" % r2_score(y_test , y_) )
```
